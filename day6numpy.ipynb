# Matirx Algebra

```python
import numpy as np

A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
```

```python
# Dot
print(np.dot(A,B))
```

```python
# matmul
A @ B
```

```python
A
```

```python
# Transpose
A.T
```

# Matrix Norms

```python
x = np.array([3,4])

np.linalg.norm(x, ord=1)   # L1 norm
```

```python

np.linalg.norm(x, ord=2)   # L2 norm (Euclidean)
```

```python

np.linalg.norm(A, 'fro')   # Frobenius norm (matrix)
```

```python
# Diterminant and Inverse Matrix
arr = np.array([[1,2,3],[4,5,6]])
arr
```

```python
# Diterminant also diterminant is only possible on squrare matrix

np.linalg.det(A)
```

```python
np.linalg.inv(A)
```

```python
#EIGENVALUES & EIGENVECTORS

# Core of PCA, covariance matrix decomposition, spectral clustering.
```

```python
eigvals, eigvecs = np.linalg.eig(A)

print("Eigenvalues:", eigvals)
print("Eigenvectors:\n", eigvecs)
```

```python
#SINGULAR VALUE DECOMPOSITION (SVD)
U, S, Vt = np.linalg.svd(A)

print("U:", U)
print("Singular Values:", S)
print("Vt:", Vt)
```

# Solving Math Problem


```python
A = np.array([[3,1],
              [1,2]])

b = np.array([9,8])

x = np.linalg.solve(A, b)
```
