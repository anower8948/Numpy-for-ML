```python
#checking the numpy version
import numpy as np
print(np.__version__)
```

```python
#creating array 

array1 = np.array([1,2,3,4,5])
array2 = np.array([[1,2,3,4],[5,6,7,8]])

print(array1)

print('\n')

print(array2)
```

```python

a_int8 = np.array([1,2,3], dtype = np.int8)
a_int16 = np.array([1,2,3], dtype= np.int16)
a_int32 = np.array([1,2,3], dtype = np.int32)
a_int64 = np.array([1,2,3], dtype=np.int64)

print(a_int8)
print(a_int16)
print(a_int32)
print(a_int64)
```

```python
print(a_int64.dtype)
```

```python
a_uint8    = np.array([1,2,3], dtype=np.uint8)
a_uint16   = np.array([1,2,3], dtype=np.uint16)
a_uint32   = np.array([1,2,3], dtype=np.uint32)
a_uint64   = np.array([1,2,3], dtype=np.uint64)


a_float16  = np.array([1,2,3], dtype=np.float16)
a_float32  = np.array([1,2,3], dtype=np.float32)
a_float64  = np.array([1,2,3], dtype=np.float64)
```

```python
print(a_uint32.dtype)
print(a_float64.dtype)
```

```python
a_complex64   = np.array([1,2,3], dtype=np.complex64)
a_complex128  = np.array([1,2,3], dtype=np.complex128)
```

```python
print(a_complex128.dtype)
```

```python
string_byte = np.array(['2','3','4','5'] , dtype='S')
string_int = np.array(['2','3','4','5'] , dtype='U')
#Unicode and Byete string
print(string_byte.dtype)
print(string_int.dtype)
```

```python
boolean_type = np.array([1,0,1,0], dtype = np.bool_)

print(boolean_type.dtype)
```

```python
obj_type = np.array([1,3,'Amir',203.4,True], dtype=object)
print(obj_type.dtype)
```
