```python
arr = np.arange(1,17,1).reshape(4,4)
print(arr.dtype)
print(arr.size)
print(arr.shape)
```

Compare memory size of int32 vs int64 arrays

```python
arr32 = np.array([1,2,3],dtype=np.int32)
arr64 = np.array([1,2,3],dtype = np.int64)\

print(arr32.nbytes)
print(arr64.nbytes)
```
