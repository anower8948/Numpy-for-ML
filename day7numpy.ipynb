```python
# Comparison: np.ones((3, 10_000_000)) vs np.arange(10_000_000)

# np.ones((3, 10_000_000))
# - Creates a 2D array with shape (3, 10,000,000)
# - All elements are 1.0
# - Default dtype: float64
# - Total elements: 30,000,000
# - Approx. memory usage: ~240 MB
# - Common use: initialization, weights, constant matrices

# np.arange(10_000_000)
# - Creates a 1D array with shape (10,000,000,)
# - Values increase sequentially from 0 to 9,999,999
# - Default dtype: int64
# - Total elements: 10,000,000
# - Approx. memory usage: ~80 MB
# - Common use: indexing, counters, sequences

# Key difference:
# - np.ones -> constant values, larger memory footprint (here)
# - np.arange -> sequential values, smaller memory footprint
```

```python
import numpy as np
import time

x = np.arange(10_000_000)

t1 = time.time()
y = [i * 2 for i in x]       # Python loop (slow)
print("Python:", time.time() - t1)

t2 = time.time()
z = x * 2                     # NumPy vectorized (fast)
print("NumPy :", time.time() - t2)
```
