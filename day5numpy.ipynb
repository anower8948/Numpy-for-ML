# Stasticial Simulation

```python
# Normal Distribution

# Mean 50, std 10

np.random.normal(50,10, size = 50)
```

```python
# Uniform distribution

np.random.uniform(0,1, size = 50)
```

```python
# coin flip... number of toss 10, head/tail probability 0.5 , number of independet exprienment 20
np.random.binomial(n=10, p=0.5, size=20)
```

```python
# Repeting

np.repeat([1,2,3,4], repeats = 2)
```

```python
# repeting rows

a = np.array([[1,2,3],[4,5,6]])

np.repeat(a,repeats = 2, axis = 0) # axis = 0 is always row....
```

```python
np.repeat(a,repeats = 2, axis = 1) # axis = 1 is always columns....
```

```python
a
```

```python
# full reps repeat

np.tile([1,2,3],reps = 3)
```

```python
np.tile([[1,2],[3,4]], reps=(2,3)) # (row_repeat , columns_repeat) (2 = 3 times for row and 3 = 4 times for column)
```
