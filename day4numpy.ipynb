```python
# Stacking Arrays
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

np.vstack([a,b])

```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

np.hstack([a,b])
```

```python
np.dstack([a, b])
```

```python
# Sort 2D Along Axis

arr = np.array([[3,1,2],
                [6,4,5]])
np.sort(arr,axis =0) # Column
```

```python
arr = np.array([[1,9,3],
                [4,2,6]])

arr.argmax()       # global index
arr.argmin()
```

```python

arr.argmax(axis=0) # per column
# First column 4(index 1) is max, second column 9(index 0) is max and last 6 ( index 1) is max so --> 1,0,1
```

```python
arr.argmax(axis=1) # per row
```
